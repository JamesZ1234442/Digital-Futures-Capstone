## Capstone EDA

```python
## Importing the modules needed

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Read in the file

df = pd.read_csv('merged.csv')
```

```python
## Checking that it has been imported correctly

df.head(20)
```

```python
## There are quite a few columns, using domain knowledge, I judged some of these to be unncessary

df.drop(columns=['Unnamed: 0', 'player_id', 'current_club_id', 'country_of_birth', 'city_of_birth', 'date_of_birth', 'foot', 'height_in_cm', 'agent_name', 'first_name', 'last_name', 'player_code', 'image_url', 'last_season', 'url', '#', 'Unnamed: 1', 'highest_market_value_in_eur'], inplace=True)
```

```python
## We have reduced the number of columns to 21

df.head(20)
```

```python
## General information of the columns

df.info()
```

```python
## We can find numerical information on the columns

df.describe()
```

```python
## we want to drop null values if they are only 1-3% of the total data

df.isnull().sum()
```

```python
## We only have 3 columns with null values

df.dropna(subset=['market_value_in_eur', 'contract_expiration_date'], inplace=True)
```

```python
## For the null values in sub_positions, it is highly likely that it is null because the player is a goalkeeper

df[df['position'] == 'Goalkeeper']
```

```python
## We want to fill these nulls

df['sub_position'].fillna('Goalkeeper', inplace=True)
```

```python
## Now we have 0 nulls in our dataset

df.isnull().sum()
```

```python
## We do not want spaces in our column names

df.rename(columns= {'Overall Score' :'overall_score', 'Potential Score':'potential_score', 'Preferred Foot':'preferred_foot', 'Ball Skills':'ball_skills'}, inplace = True)
```

```python
## visualise the correlations

plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
## let us visualise the datas against each other
## the target variable here is the market value column
## age against market value

sns.barplot(x = df['Age'], y=df['market_value_in_eur'], ci=None)
#plt.savefig('age_against_value')
plt.show()
```

```python
## overall score against market value

sns.scatterplot(x=df['overall_score'], y=df['market_value_in_eur'])
plt.show()
```

```python
## it seems like an exponential increase, so let us log the market value

sns.scatterplot(x=df['overall_score'], y=np.log(df['market_value_in_eur']))
plt.show()
```

```python
## Indeed, it seems like we do have a linear relationship between the log(market value) and the overall score
## however, with a few outliers
## this can be explained through the high potential scores assigned to some very talented players who are young

sns.scatterplot(x=df['potential_score'], y=df['market_value_in_eur'])
#plt.savefig('potential_against_values')
plt.show()
```

```python
## Similarly, the increase seems to be exponential, which is expected

sns.scatterplot(x=df['potential_score'], y=np.log(df['market_value_in_eur']))
plt.show()
```

```python
## Realistically, we expect forwards to be valued more than defenders
## lets see if our data reflects this

sns.barplot(x=df['position'], y=df['market_value_in_eur'], ci = None)
#plt.savefig('position_against_value')
plt.show()
```

```python
## we expect Premier league players to be valued more as Premier league is often considered to be the best league

order = df.groupby('current_club_domestic_competition_id')['market_value_in_eur'].mean().sort_values(ascending = False).index
sns.barplot(x=df['current_club_domestic_competition_id'], y=df['market_value_in_eur'], order = order, ci = None)
#plt.savefig('comp_against_value')
plt.show()
```

```python
## there are a lot of countries which the players play for
## let us take the top 15 countries by count

df['country_of_citizenship'].value_counts()
```

```python
## there are probably a faster way to do this...

country_list = ['France', 'England', 'Germany', 'Brazil', 'Netherlands', 'Portugal', 'Italy', 'Belgium', 'Denmark', 'Turkey', 'Scotland', 'Argentina', 'Sweden', 'Norway']
df_top15 = df[df['country_of_citizenship'] == 'Spain'].copy()
for countries in country_list:
    country_df = df[df['country_of_citizenship'] == countries]
    df_top15 = pd.concat([df_top15, country_df])
order = df_top15.groupby('country_of_citizenship')['market_value_in_eur'].mean().sort_values(ascending=False).index
plt.figure(figsize=(10,10))
sns.barplot(x=df_top15['country_of_citizenship'], y=df_top15['market_value_in_eur'], ci = None, order = order)
plt.savefig('countries_against_values')
plt.show()
```

```python
## We can look at which current on average have the highest valued players

df.groupby('current_club_name')['market_value_in_eur'].mean()
```

```python
## Will preferred foot impact the market value?

sns.barplot(x=df['preferred_foot'], y=df['market_value_in_eur'])
plt.show()
```
